```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import math
import warnings
warnings.filterwarnings( "ignore" )

col_names = ['x1','x2','result']
#load dataset
df = pd.read_csv("Student-University.csv", names=col_names)
print(df)
```

```python
from sklearn.preprocessing import StandardScaler
StandardScaler().fit_transform(df)
```

```python
X = df.values[:,0:2]
```

```python
Y = df.values[:,2]
```

```python
# Splitting dataset into train and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
X_train
```

```python
x1 = X_train[:,0]
x2 = X_train[:,1]
```

```python
b0=0.0
b1=0.0
b2=0.0
alpha = 0.001
#print(Y_train)
#prediction = 1/(1+np.exp(-y))
#print(np.round(prediction))
#print(Y_train) #predicted and training values are different
epoch = 10000
x=1
while(epoch>0):
    for i in range(len(X_train)):
        prediction = 1/(1+np.exp(-(b0+b1*x1[i]+b2*x2[i])))
        # Update b0,b1,b2
        b0=b0+alpha*(Y_train[i]-prediction)*prediction*(1-prediction)*x
        b1=b1+alpha*(Y_train[i]-prediction)*prediction*(1-prediction)*x1[i]
        b2=b2+alpha*(Y_train[i]-prediction)*prediction*(1-prediction)*x2[i]
    epoch=epoch-1
```

```python
b0,b1,b2
```

```python
x3=X_test[:,0]
x4=X_test[:,1]
print(Y_test)
final_prediction=[]
for i in range(len(X_test)):
    y_final = b0 + b1*x3[i] + b2*x4[i]
    prediction = 1/(1+np.exp(-y_final))
    final_prediction.append(np.ceil(prediction))
    #print(int(final_prediction))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,final_prediction)
```
